## **3. Feature-Target Isolation & Stratified Data Splitting**

### **Purpose:**

The purpose of this step is to partition our dataset into distinct subsets for model training and unbiased evaluation. First, we separate our independent predictive features ($X$) from the dependent target label ($y$). Then, we split the samples into training ($80\%$) and testing ($20\%$) subsets while enforcing class proportion balance across both splits to prevent training bias.

### **Key Points**

* **Data Leakage Mitigation:** Splitting the data prior to scaling or feature modifications ensures that the model evaluation phase remains completely uninfluenced by information from the training domain.

* **Class Balance Preserved:** Utilizing stratified splitting is crucial for binary classification tasks, as it guarantees that the ratio of spam to ham remains identical in both the training and testing segments.

* **Reproducible Partitioning:** Setting a constant pseudorandom seed fixes the data shuffling pattern, ensuring consistency across separate validation runs.

### **Code Architecture Breakdown**

* `df.drop(columns=['is_spam'])`: Isolates the core independent features (`num_links`, `num_words`, `has_offer`, `sender_score`, `all_caps`) into matrix $X$.

* `test_size=0.2`: Allocates $20\%$ of the total data volume ($4,000$ samples) for validation, leaving $80\%$ ($16,000$ samples) for optimization.

* `stratify=y`: Controls the selection mechanism to replicate the global distribution of the binary classification labels within both localized splits.

```python
# ============================
# 1. FEATURE-TARGET ISOLATION
# ============================
# Extract independent features by dropping the ground-truth prediction column
X = df.drop(columns=['is_spam'])

# Isolate the dependent binary class label target vector (0: Ham, 1: Spam)
y = df['is_spam']

# ========================================================
# 2. STRATIFIED VALIDATION SPLITTING (80% Train, 20% Test)
# ========================================================
# Stratify against target 'y' to lock balanced classification ratios in both sets
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,       # Dedicate exactly 20% of dataset samples to testing evaluation
    random_state=42,     # Fixed random seed to enforce reproducible shuffle boundaries
    stratify=y           # Enforce equal class distributions between splits to avoid bias
)

# ==============================
# 3. DIMENSIONAL VERIFICATION
# ==============================
# Confirm accurate matrix alignment before passing arrays to tensors
print(f"Training features shape: {X_train.shape}")
print(f"Testing features shape: {X_test.shape}")
```

## **4. Feature Scaling & Data Leakage Prevention**

### **Purpose:**

The purpose of this step is to standardize our independent numeric features using a scaling transformation. Because features like word count span large ranges compared to binary flags, scaling ensures that all parameters contribute equally to the gradient descent optimization of our neural network, preventing features with large values from dominating the weight updates.

### **Key Points**
* **Gradient Optimization Enhancement:** Centering values around a mean of zero and standard deviation of one prevents vanishing or exploding gradients during training.

* **Strict Leakage Prevention:** Fitting the scaler exclusively on the training array prevents the model from gaining prior knowledge about the distribution of the test partition.

* **Consistency in Evaluation:** Applying the exact same learned scaling parameters to the test set guarantees a stable mathematical translation for validation.

### **Code Architecture Breakdown**

* `StandardScaler()`: Instantiates the transformation object that handles standard score normalization.

* `scaler.fit_transform(X_train)`: Calculates the column-wise mean and variance values from the training data and converts the raw inputs into scaled representations.

* `scaler.transform(X_test)`: Maps the test dataset features using the pre-calculated training weights without recalculating the mean and variance.

```python
# Initialize the standard score normalization scaler
scaler = StandardScaler()

# Fit parameters on training data and transform it
X_train_scaled = scaler.fit_transform(X_train)

# Transform test data using training parameters to prevent data leakage
X_test_scaled = scaler.transform(X_test)

# Verify the scaling conversion output structure
print("First processed row of training data:\n", X_train_scaled[0])
```

# **🧠 Phase 2: Engine 1 Framework (Deep MLP Core)**

## **5. Deep Learning Model Architecture & Compilation**

### **Purpose:**

The purpose of this step is to design and build a Multi-Layer Perceptron (MLP) Neural Network for binary spam classification. The architecture utilizes progressive layer sizing alongside optimization layers to capture complex interactions among the five extracted email and message features.

### **Key Points**

* **Regularization via Dropout:** Introducing Dropout layers across hidden nodes prevents overfitting by forcing the network to learn redundant representations.

* **Internal Covariate Shift Control:** Integrating Batch Normalization layers stabilizes and accelerates training by normalizing layer inputs after every dense mathematical transformation.

* **Probability Mapping Function:** The final single-node layer utilizes a Sigmoid activation function to constrain predictive scores between $0.0$ and $1.0$, representing structural threat probability.

### **Code Architecture Breakdown**

* `input_shape=(input_dim,)`: Sets the receiving boundary of the input array to accept our $5$ standardized metrics.

* `loss='binary_crossentropy'`: Selects the standard loss function for two-class target tasks to evaluate discrepancy distances between target classes and prediction probabilities.

* `model.summary()`: Generates a complete breakdown displaying layer topology, output shape dimensions, and internal parameter weight calculations ($12,033$ total parameters).

```python
# Extract the total count of predictive features (5 in our dataset)
input_dim = X_train_scaled.shape[1]

# Construct a linear stack of deep neural network layers
model = Sequential([
    # Input Layer paired with the First Hidden Layer
    Dense(128, activation='relu', input_shape=(input_dim,)),
    BatchNormalization(),
    Dropout(0.3), # Randomly drops 30% of node weights to prevent overfitting

    # Second Hidden Layer
    Dense(64, activation='relu'),
    BatchNormalization(),
    Dropout(0.3),

    # Third Hidden Layer
    Dense(32, activation='relu'),
    BatchNormalization(),
    Dropout(0.2),

    # Output Layer utilizing Sigmoid activation for binary probability mapping
    Dense(1, activation='sigmoid')
])

# Configure the optimization training parameters of the model
model.compile(
    optimizer='adam',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

# Output a structural breakdown of the complete network topology
model.summary()
```

## **6. Hyperparameter Optimization & Advanced Neural Topology**

### **Purpose:**

The purpose of this step is to implement a highly optimized, deeper neural network configuration designed to scale up structural patterns. By integrating advanced mathematical layers such as LeakyReLU alongside custom optimizer configurations, the model addresses common deep-learning bottlenecks like node deadlocks, thereby squeezing maximum precision out of our classification features.

### **Key Points**

* **Preventing Dead Neurons:** Using `LeakyReLU` instead of standard `ReLU` prevents the "dying ReLU" problem. It does this by maintaining a small gradient slope ($\alpha = 0.1$) for negative values rather than zeroing them out completely.

* **Scaled Model Capacity:** Expanding the hidden architecture to $4$ sequential layers ($256 \rightarrow 128 \rightarrow 64 \rightarrow 32$) allows the system to construct higher-level abstractions from raw data signals.

* **Controlled Weight Initialization:** Initializing a explicit instance of the `Adam` optimizer provides strict control over the base learning rate parameter ($0.001$) for precision tuning.

### **Code Architecture Breakdown**

* `LeakyReLU(alpha=0.1)`: Allows a $10\%$ continuous pass-through of negative tensor weights to sustain error propagation updates.

* `tf.keras.optimizers.Adam()`: Instantiates the foundational gradient optimization engine with customizable parameter steps.

* `advanced_model.summary()`: Generates a topological report detailing structural layers and mathematical parameter mappings ($59,329$ total parameters).

```python
# Import advanced activation configurations and optimization components
from tensorflow.keras.layers import LeakyReLU
from tensorflow.keras.callbacks import ReduceLROnPlateau

# =====================================================================
# 1. ADVANCED DEEP NEURAL NETWORK TOPOLOGY
# =====================================================================
advanced_model = Sequential([
    # First Hidden Layer with expanded node capacity
    Dense(256, input_shape=(input_dim,)),
    LeakyReLU(alpha=0.1), # Prevents zero-gradient blocks for negative weights
    BatchNormalization(),
    Dropout(0.4),

    # Second Hidden Layer
    Dense(128),
    LeakyReLU(alpha=0.1),
    BatchNormalization(),
    Dropout(0.4),

    # Third Hidden Layer
    Dense(64),
    LeakyReLU(alpha=0.1),
    BatchNormalization(),
    Dropout(0.3),

    # Fourth Hidden Layer
    Dense(32),
    LeakyReLU(alpha=0.1),
    BatchNormalization(),
    Dropout(0.2),

    # Binary probability mapping output node
    Dense(1, activation='sigmoid')
])

# =====================================================================
# 2. OPTIMIZER SPECIFICATION & COMPILATION
# =====================================================================
# Instantiate an isolated Adam optimizer with explicit initial base step scales
custom_adam = tf.keras.optimizers.Adam(learning_rate=0.001)

# Compile the advanced model structural parameters
advanced_model.compile(
    optimizer=custom_adam,
    loss='binary_crossentropy',
    metrics=['accuracy']
)

# Output detailed computational breakdown characteristics of the advanced net
advanced_model.summary()
```

## **7. Advanced Optimization Training with Dynamic Learning Rate Scheduling**

### **Purpose:**

The purpose of this step is to execute the optimization loop for our deeper, advanced neural network. To ensure efficient convergence, we introduce a dual-callback strategy: `EarlyStopping` prevents overfitting, while `ReduceLROnPlateau` dynamically shrinks the optimization step size when loss stagnation is detected, helping the model settle into a more precise global minimum.

### **Key Points**

* **Adaptive Gradient Descent Steps:** Incorporating `ReduceLROnPlateau` dynamically modifies the active learning speed based on metric tracking, reducing it by $50\%$ (`factor=0.5`) if the validation loss stagnates for $2$ consecutive epochs.

* **Enhanced Model Generalization:** Shifting the processing configuration to a smaller structural batch size (`batch_size=32`) introduces beneficial stochastic noise into weight updates, preventing the deep network from getting stuck in sharp, unoptimized local minima.

* **Optimized Grace Period Coordination:** Setting a patience threshold of $6$ epochs for early stopping ensures the optimizer has enough time to evaluate performance improvements after a learning rate reduction occurs.

### **Code Architecture Breakdown**

* `lr_scheduler = ReduceLROnPlateau(...)`: Instantiates the automated scheduling callback engine to monitor performance bottlenecks and decay learning speed dynamically.

* `factor=0.5, patience=2`: Multiplies the learning rate by $0.5$ if validation loss fails to decrease for two consecutive checks.

* `callbacks=[early_stopping, lr_scheduler]`: Binds both tracking engines directly into the runtime sequence execution of the network optimization step.

```python
# Configure a baseline early stopping safety monitor with a patience of 6 epochs
early_stopping = EarlyStopping(
    monitor='val_loss',
    patience=6,
    restore_best_weights=True
)

# Initialize an adaptive learning rate scheduler callback engine
lr_scheduler = ReduceLROnPlateau(
    monitor='val_loss',
    factor=0.5,            # Reduce the learning rate by half when a plateau occurs
    patience=2,            # Number of epochs to wait before decaying the step scale
    min_lr=1e-6,           # Set the absolute lowest floor bound for the learning rate
    verbose=1              # Trigger explicit console alerts when learning rate adjustment occurs
)

# Launch the training optimization cycle on our advanced deep neural network
adv_history = advanced_model.fit(
    X_train_scaled,
    y_train,
    epochs=50,             # Set the maximum possible learning epochs upper limit
    batch_size=32,         # Utilize smaller batch blocks to optimize model generalization
    validation_split=0.2,  # Isolate 20% of data for intermediate validation updates
    callbacks=[early_stopping, lr_scheduler], # Register both tracking callbacks
    verbose=1              # Print real-time epoch telemetry parameters
)
```

## **8. Neural Network Training & Early Stopping Optimization**

### **Purpose:**
The purpose of this step is to execute the actual training process of our deep neural network. By feeding the scaled training data through the model in mini-batches across multiple iterative cycles (epochs), the network systematically updates its internal weights to minimize prediction errors on spam classification.

### **Why Do We Use These Specific Configurations?**

* **Why `EarlyStopping`?** Without a monitor, neural networks will continue training for all specified epochs, causing them to eventually memorize noise in the training set and break down on unseen data. Early stopping halts training the moment validation performance declines, preserving ideal model generalization.

* **Why `validation_split=0.2`?** It carves out a clean $20\%$ portion of the training data to act as an immediate test mirror during training. This lets us monitor performance on unseen data at the end of every single epoch without touching our final test data.

* **Why `patience=5`?** It acts as a safety buffer. Optimization paths fluctuate, so a patience value of $5$ tells the engine not to stop immediately on a single bad epoch, but to wait and see if the loss continues to degrade for five consecutive steps before concluding.

* **Why `restore_best_weights=True`?** When training stops, the network weights are stuck at the final unoptimized epoch where performance was worst. Enabling this parameter automatically rolls back the entire network model to the exact epoch that recorded the lowest absolute validation loss.

### **Key Points**
* **Iterative Learning Rate:** Processes the optimization over a maximum bound of $50$ epochs, using a standard mathematical batch chunk size of $64$ samples.

* **Overfitting Prevention:** Combines dropout regularization with live validation tracking to actively identify data over-fitting thresholds.

### **Code Architecture Breakdown**

* `monitor='val_loss'`: Instructs the callback utility to continuously analyze the validation loss trajectory to determine early cutoff points.

* `model.fit()`: Launches the main computational execution loop that updates neural weight tensors across memory matrices.

```python
# Configure early stopping callback to track validation loss trends
early_stopping = EarlyStopping(
    monitor='val_loss',
    patience=5,
    restore_best_weights=True
)

# Launch the neural network training loop on our scaled dataset
history = model.fit(
    X_train_scaled,
    y_train,
    epochs=50,                  # Maximum number of complete learning cycles
    batch_size=64,              # Number of training samples processed per gradient update
    validation_split=0.2,       # Hold out 20% of training data for dynamic performance checks
    callbacks=[early_stopping], # Inject the early stopping optimization monitor
    verbose=1                   # Display step-by-step progress bars and metrics for each epoch
)
```

## **11: Advanced 1D-CNN Network Topology, Training, and Serialization Core**

### **Purpose:**

The objective of this module is to construct, optimize, and serialize our secondary deep learning engine using a 1D Convolutional Neural Network (1D-CNN). While traditional dense networks treat input components independently, this engine applies specialized convolutional kernels to scan the input features sequentially. This architectural strategy allows the pipeline to learn hidden spatial relationships and structural dependencies across scaled feature metrics, forming a unique classification perspective that significantly enhances our ensemble's final consensus voting reliability.

### **Key Points**

* **Dimensional Reshaping Processing:** Utilizing np.expand_dims converts raw standardized input matrices from a flat 2D structural profile into a precise 3D sequential format (samples, features, channels) structurally demanded by 1D convolutional network layers.

* **Feature Extraction Kernels:** Integrates specialized 1D sliding spatial extraction layers (Conv1D) with growing filter counts (32 moving to 64) to capture complex, localized correlation structures.

* **Regularization Matrix Safeguards:** Alternates sequential convolution passes with robust structural layers, including BatchNormalization to maintain gradient stability and Dropout matrices to completely suppress overfitting hazards.

# **🤖 Phase 5: Real-Time NLP Ingestion Engine**

## **14. Automated Natural Language Feature Extraction & Multi-Engine Processing**

### **Purpose:**

The purpose of this step is to build an automated, end-to-end processing pipeline that transforms raw, unformatted string text into clean numerical features. Instead of requiring pre-calculated dataset metrics, this pipeline utilizes regular expressions (Regex) and text processing logic to parse incoming text strings in real-time, scale the extracted features, and feed them into the ensemble decision matrix.

### **Key Points**

* **Dynamic Pattern Matching:** Uses Regular Expressions (`re`) to dynamically extract technical structural footprints like hyperlink density, shouting markers (`ALL_CAPS`), and aggressive promotional keywords.

* **Proportional Flag Thresholding:** Implements an evaluation check where the shouting flag triggers only if capitalized keywords make up more than $40\%$ of the total word count, reducing false positives.

* **Autonomous Ingest Execution:** Combines raw string analysis directly with pre-trained storage checkpoints to instantly generate diagnostic security ratings without human intervention.

### **Code Architecture Breakdown**

* `re.findall(r'...')`: Matches alphanumeric patterns against a standard web URL signature to calculate link totals.

* `any(word in raw_text.lower() ...)`: Normalizes casing differences and checks the message body against a security blacklist of high-frequency spam triggers.

* `final_prob = (prob_mlp + prob_cnn) / 2`: Computes the definitive arithmetic mean threshold boundary from both independent deep learning models.

```python
import re
import numpy as np
import tensorflow as tf
import joblib
import warnings

# Suppress operational framework notices to keep prediction reporting clean
warnings.filterwarnings("ignore")

# =====================================================================
# 1. INITIALIZE MEMORY CHECKPOINTS & SERIALIZED MODELS
# =====================================================================
try:
    mlp_model = tf.keras.models.load_model('advanced_spam_detector_model.h5', compile=False)
    cnn_model = tf.keras.models.load_model('cnn_spam_detector_model.h5', compile=False)
    data_scaler = joblib.load('feature_scaler.pkl')
except Exception as e:
    print("Inference engine loading error:", e)

# =====================================================================
# 2. AUTOMATIC NATURAL LANGUAGE FEATURE EXTRACTION PIPELINE
# =====================================================================
def automatic_text_feature_extractor(raw_text, sender_score_input):
    try:
        # Split text into elements to compute absolute token length
        words_list = raw_text.split()
        num_words = len(words_list) if len(words_list) > 0 else 1

        # Isolate and count instances of embedded web hyperlinks using Regex
        links_found = re.findall(r'(https?://[^\s]+|www\.[^\s]+)', raw_text)
        num_links = len(links_found)

        # Screen text body for promotional keywords or security alerts
        promo_keywords = ['offer', 'prize', 'win', 'free', 'cash', 'bonus', 'discount', 'click here', 'lottery', 'urgent']
        has_offer = 1 if any(word in raw_text.lower() for word in promo_keywords) else 0

        # Trigger shouting flag if uppercase words exceed a 40% density threshold
        caps_words = [word for word in words_list if word.isupper() and len(word) > 1]
        all_caps = 1 if (len(caps_words) / num_words) > 0.4 else 0

        sender_score = float(sender_score_input)

        # =====================================================================
        # 3. VECTORIZATION & DUAL-ENGINE ENSEMBLE VOTING
        # =====================================================================
        # Reshape extracted attributes into a standard 2D vector for scaling
        features = np.array([[num_links, num_words, has_offer, sender_score, all_caps]], dtype=np.float32)
        scaled_features = data_scaler.transform(features)

        # Engine 1: Extract prediction probability using the Multi-Layer Perceptron
        prob_mlp = mlp_model.predict(scaled_features, verbose=0)[0][0]

        # Engine 2: Reshape features to 3D and extract prediction via 1D-CNN
        scaled_features_cnn = np.expand_dims(scaled_features, axis=-1)
        prob_cnn = cnn_model.predict(scaled_features_cnn, verbose=0)[0][0]

        # Calculate final consensus rating by averaging engine scores
        final_prob = (prob_mlp + prob_cnn) / 2

        # =====================================================================
        # 4. DIAGNOSTIC FORENSICS REPORT GENERATION
        # =====================================================================
        status = "🚨 CRITICAL SPAM / PHISHING DETECTED" if final_prob > 0.5 else "✅ HAM / LEGITIMATE COMMUNICATION"
        confidence = final_prob * 100 if final_prob > 0.5 else (1 - final_prob) * 100
        action = "QUARANTINE MESSAGE IMMEDIATELY" if final_prob > 0.5 else "DELIVER TO INBOX"

        report = (
            f"============================================================\n"
            f"      🌌 COGNITIVE UNIFIED TEXT INTELLIGENCE CORE         \n"
            f"============================================================\n"
            f" ANALYSIS STATUS    : {status}\n"
            f" SYSTEM CONFIDENCE  : {confidence:.2f}%\n"
            f" SUGGESTED ACTION   : {action}\n"
            f"------------------------------------------------------------\n"
            f"              📊 AUTO-EXTRACTED PAYLOAD METRICS            \n"
            f"------------------------------------------------------------\n"
            f" • Extracted Links  : {num_links} detected\n"
            f" • Word Density     : {num_words} total words parsed\n"
            f" • Promo Keyword    : {'Yes (Triggered Flag)' if has_offer == 1 else 'None'}\n"
            f" • Shouting Flag    : {'Yes (High Density Caps)' if all_caps == 1 else 'No'}\n"
            f" • Sender Risk Base : {sender_score:.2f}/1.00\n"
            f"------------------------------------------------------------\n"
            f"              🎯 BACKEND LAYER VOTING SPLIT                \n"
            f"------------------------------------------------------------\n"
            f" 🧠 Engine 1 (Multi-Layer Perceptron) -> Weight Prob: {prob_mlp:.4f}\n"
            f" 🧬 Engine 2 (1D Convolutional Net)   -> Weight Prob: {prob_cnn:.4f}\n"
            f"============================================================"
        )
        return report

    except Exception as err:
        return f"❌ Parsing Breakdown Failure: {str(err)}"
```

## **17. Hybrid Dual-Engine Inference & Prediction Consensus Core**

### **Purpose:**

The purpose of this step is to implement a hybrid ensemble decision engine. Instead of relying on a single model architecture, this advanced scanner feeds input data into two structurally distinct deep learning networks simultaneously: a Multi-Layer Perceptron (MLP) and a 1D Convolutional Neural Network (1D-CNN). By averaging their independent prediction probabilities, the system creates a voting consensus that yields a more robust and dependable final decision.

### **Key Points**

* **Multi-Architecture Ensemble:** Combining a feature-focused network (MLP) with a sequential spatial pattern network (1D-CNN) reduces individual model biases and increases defense reliability.

* **Dynamic Dimensional Reshaping:** Automatically expands the standardized feature arrays into a 3D tensor format (`[samples, features, channels]`) on the fly, satisfying the structural shape requirements of the 1D-CNN layer.

* **Unified Telemetry Logging:** Merges the output parameters, structural confidence scores, and safety routing actions into a single, clean text-based report for real-time monitoring.

### **Code Architecture Breakdown**

* `tf.keras.models.load_model(..., compile=False)`: Instantiates both pre-trained network frameworks from storage without requiring optimization configurations.

* `np.expand_dims(scaled_inputs, axis=-1)`: Transforms the 2D input array into a 3D shape, adding a channel dimension to fit the 1D-CNN core.

* `final_prob = (prob_mlp + prob_cnn) / 2`: Calculates the arithmetic mean of both model probabilities to establish an ensemble classification decision boundary.

```python
import numpy as np
import tensorflow as tf
import joblib
import warnings

# Suppress operational notices to keep output reports clean
warnings.filterwarnings("ignore")

# =======================================
# 1. SERIALIZED COMPONENT INGESTION CORE
# =======================================
try:
    # Load both models without compilation parameters for faster inference setup
    mlp_model = tf.keras.models.load_model('advanced_spam_detector_model.h5', compile=False)
    cnn_model = tf.keras.models.load_model('cnn_spam_detector_model.h5', compile=False)
    data_scaler = joblib.load('feature_scaler.pkl')
except Exception as e:
    print("Component Load Error:", e)

# ===========================================
# 2. DUAL-ENGINE ENSEMBLE COGNITIVE SCANNER
# ===========================================
def dual_engine_cognitive_scanner(num_links, num_words, has_offer, sender_score, all_caps):
    try:
        # Preprocess and format raw inputs into a standard float32 numpy array
        inputs = np.array([[float(num_links), float(num_words), float(has_offer), float(sender_score), float(all_caps)]], dtype=np.float32)
        scaled_inputs = data_scaler.transform(inputs)

        # Engine 1: Compute Multi-Layer Perceptron (MLP) network prediction
        prob_mlp = mlp_model.predict(scaled_inputs, verbose=0)[0][0]
        status_mlp = "🚨 SPAM" if prob_mlp > 0.5 else "✅ HAM"
        conf_mlp = prob_mlp * 100 if prob_mlp > 0.5 else (1 - prob_mlp) * 100

        # Engine 2: Reshape input to 3D and compute 1D-CNN network prediction
        scaled_inputs_cnn = np.expand_dims(scaled_inputs, axis=-1)
        prob_cnn = cnn_model.predict(scaled_inputs_cnn, verbose=0)[0][0]
        status_cnn = "🚨 SPAM" if prob_cnn > 0.5 else "✅ HAM"
        conf_cnn = prob_cnn * 100 if prob_cnn > 0.5 else (1 - prob_cnn) * 100

        # Ensemble Logic: Compute the voting consensus score
        final_prob = (prob_mlp + prob_cnn) / 2
        final_status = "🚨 CRITICAL SPAM DETECTED" if final_prob > 0.5 else "✅ SECURE / HAM EMAIL"
        security_action = "IMMEDIATE QUARANTINE" if final_prob > 0.5 else "SAFE TO DELIVER"

        # 3. Generate the comprehensive multi-engine diagnostic text report
        report = (
            f"============================================================\n"
            f"         🛡️ MULTI-ENGINE COGNITIVE AI DEEP LEARNING REPORT  \n"
            f"============================================================\n"
            f" FINAL CONSENSUS   : {final_status}\n"
            f" SECURITY ACTION   : {security_action}\n"
            f"------------------------------------------------------------\n"
            f"               🎯 NEURAL NETWORK ENGINE COMPARISON          \n"
            f"------------------------------------------------------------\n"
            f" 🧠 Engine 1 (Multi-Layer Perceptron) -> {status_mlp} ({conf_mlp:.2f}% Confidence)\n"
            f" 🧬 Engine 2 (1D-Convolutional Net)   -> {status_cnn} ({conf_cnn:.2f}% Confidence)\n"
            f"------------------------------------------------------------\n"
            f"                📊 LIVE TELEMETRY FEATURE INPUTS             \n"
            f"------------------------------------------------------------\n"
            f" • Link Density Checked   : {int(num_links)} links parsed\n"
            f" • Scanned Token Capacity : {int(num_words)} total words\n"
            f" • Reputation Matrix Base : {float(sender_score):.2f}/1.00\n"
            f"============================================================"
        )
        return report

    except Exception as err:
        return f"❌ System Error Pipeline Failed: {str(err)}"
```
